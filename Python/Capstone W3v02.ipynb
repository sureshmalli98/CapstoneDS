```python
#Scraping to get borough list in Toronto Create a soup object and filter the required columns after locating table

page = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(page.content, 'html.parser')

table_contents=[]
table=soup.find('table')
#print('table',table)
for row in table.findAll('td'):
    cell = {}
    #print('row',row)
    #print('row.span',row.span,'1')
    if row.span.text=='Not assigned':
        pass
    else:
        cell['PostalCode'] = row.p.text[:3]
        cell['Borough'] = (row.span.text).split('(')[0]
        cell['Neighborhood'] = (((((row.span.text).split('(')[1]).strip(')')).replace(' /',',')).replace(')',' ')).strip(' ')
        table_contents.append(cell)

df=pd.DataFrame(table_contents)
#df['Borough']=df['Borough'].replace({'Downtown TorontoStn A PO Boxes25 The Esplanade':'Downtown Toronto Stn A',
#                                  df[df['column_name'] == ''].index                                          'MississaugaCanada Post Gateway Processing Centre':'Mississauga'})
```

```python

geodata = pd.read_csv("https://cocl.us/Geospatial_data")
#print("Data read into dataframe!")
geodata = geodata.rename(columns={"Postal Code": "PostalCode"})


df_postcode = df.groupby(['PostalCode', 'Borough'])['Neighborhood'].apply(', '.join).reset_index()
df_postcode.head()
toronto_data = pd.merge(df_postcode,geodata,left_on='PostalCode',right_on='PostalCode',left_index=False,right_index=False)

print('data',toronto_data.head())
toronto_data.shape
toronto_data.isnull()

```
